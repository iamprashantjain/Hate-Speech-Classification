### Project Introduction - End to End NLP Project Implimentation - Text (Hate Speech) Classification

- Its a process of identifying and categorizing speech of language that may be considered offensive/harmful
- Like in chatgpt: If we write "how to kill someone?" It wont reply!!
- In backend, Its performing hate speech classification & checking what kind of prompt is given.
- Except chatgpt, This is being used in various applications.
- Also SORA, which can generate videos on the basis of text input, but in order to regularize it to not generate videos on hate speech we need a solution
- Pipeline remain similar for Hatespeech classification, Sentimental Analysis etc

**How**

**End to End Project: NLP Pipelines**
- breaking down the problem into several sub problems (components)

    1. Data gathering : webscraping, fetching from DB/API, prepare own data(data augmentation in case of less data like replace with synonyms, biagram flip, back translate, adding additional noise)
    2. Text preparation : 
        + `basic preprocessing`: tokenization (sentence level, word level), remove stop words (in some cases we use it like in chatgpt we use stopwords also), stemming/lamatization (bringing word to root form like play, playing, played to "play"), removing punctuations, change text to lower case 
        + `advanced preprocessing`: POS tags & parsing to understanding the grammatical context, coreference resolution

        

    3. Feature Engineering: text vectorization (converting text to numbers using BOW, TFiDF, OHE, Word2Vec, embedding models), label data encoding (hate & no hate)
    4. Modelling: Huristic approach, ML(Naivebayes or Randomforst), DL(RNN reccural nueral network), Cloud API
    5. Evaluation: Intrinsive (accuracy score, f1 score, precision, recall etc) & Extrensive evaluation (taking live feedback from users)

    

    6. Deployment: To cloud
    7. Monitoring/Retraining
